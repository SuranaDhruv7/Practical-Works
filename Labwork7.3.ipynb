## Pivoting

1. Create a DataFrame containing sales data for a store with columns: 'Date', 'Category', 'Sales'. Pivot the DataFrame to show the total sales for each category by date.

```python
import pandas as pd
# Create a Dictionary
data = {'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02', '2024-01-03', '2024-01-03', '2022-01-20', '2024-05-21', '2022-01-20', '2025-09-23'],
        'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing', 'Branding', 'Material', 'Branding', 'Material'],
        'Sales': [100, 50, 150, 75, 120, 60, 25, 40, 50, 10]}

# Create a DataFrame
df1 = pd.DataFrame(data)
print(df1)
print("-------------------\n")
# Pivot the DataFrame
pivot_df = df1.pivot_table(index='Date', columns='Category', values='Sales', aggfunc='sum')
pivot_df
```

2. Create a DataFrame with columns: 'Region', 'Product', 'Revenue'. Pivot the DataFrame to calculate the average revenue for each product in different regions.

```python
data = {'Region': ['East', 'East', 'West', 'West', 'North', 'North'],
        'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Revenue': [100, 150, 120, 180, 90, 110]}

# Create DataFrame
df2 = pd.DataFrame(data)
print(df2)
print("-------------------\n")
# Pivot the DataFrame
pivot_df = df2.pivot_table(index='Region', columns='Product', values='Revenue', aggfunc='mean')
pivot_df
```

3. Using a given DataFrame, pivot the data to display the count of items sold based on 'Store' and 'Year'.

```python
df2 = pd.DataFrame({
    'Store': ['Store A', 'Store B', 'Store A', 'Store B', 'Store A', 'Storoe B']
    , 'Year': ['2023', '2023', '2022', '2022', '2021', '2021']
    , 'Item': ['Item A', 'Item B', 'Item A', 'Item B', 'Item A', 'Item B']
    , 'Items_sold': [10, 15, 5, 20, 8, 12]
})
print(df2)
print("-------------------\n")
pivot_df = df2.pivot_table(index='Store', columns='Year', values='Items_sold', aggfunc='sum')
pivot_df
```

## Sorting & Aggregation

4. Create a DataFrame with columns: 'Name', 'Age', 'Score'. Sort the DataFrame by 'Score' in descending order.

```python
df3 = pd.DataFrame({
    'Name' : ['Dev', 'Dhruv', 'Sarju', 'Namya']
    ,'Age' : [10, 15, 20, 25]
    ,'Score' : [85, 90, 75, 95]
})
print(df3)
print("-------------------\n")
sorted_df = df3.sort_values(by='Score', ascending=False)
sorted_df
```

## Groupby() & Transform()

10. Create a DataFrame with columns: 'Department', 'Employee', 'Salary'. Group the data by 'Department' and calculate the average salary for each department.

```python
df9 = pd.DataFrame({
    'Department': ['HR', 'Sales', 'HR', 'Sales']
    ,'Employee': ['Dev', 'Dhruv', 'Sarju', 'Namya']
    ,'Salary': [550000, 6600000, 7700000, 4500000]
})
df9
df9.groupby('Department')['Salary'].mean()
```

11. Create a DataFrame with sales data (Store, Item, Sales). Group the data by 'Item' and use `transform()` to calculate the percentage contribution of each sale.

```python
df10 = pd.DataFrame({
    'Store': ['Store A', 'Store B', 'Store A', 'Store B', 'Store A', 'Storere B']
    ,'Item': ['Item A', 'Item B', 'Item A', 'Item B', 'Item A', 'Item B']
    ,'Sales': [100, 150, 120, 130, 363, 120]
})
df10
df10.groupby('Item')['Sales'].transform(lambda x: x / x.sum())
```
