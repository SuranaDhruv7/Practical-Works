
## Task 1: Introduction to Matplotlib

1. Import the Matplotlib library and display its version.
2. Create a basic line plot using a list of integers.
3. Add titles and labels to the plot (title, x-axis, y-axis).


```python
import matplotlib.pyplot as plt
import matplotlib
print(matplotlib.__version__)
```

```python
line_plot = [1, 2, 3, 4, 5]
plt.plot(line_plot)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```


## Task 2: Basic Charts

1. Create a bar chart showing the sales of various items in a store.
2. Plot a line chart showing temperature changes over a week.
3. Create a scatter plot with random data for x and y coordinates.
4. Visualize a pie chart representing time spent on daily activities like work, sleep, and leisure.


```python
bar_plot = ["Soap", "Cream", "Perfume", "Ittar"]
sales = [100, 200, 150, 300]
plt.bar(bar_plot, sales)
plt.title("Bar Plot")
plt.xlabel("Items")
plt.ylabel("Sales")
plt.show()
```

```python
line_plot = [23, 1, -10, 30]
weeks = [1, 2, 3, 4]
plt.plot(weeks, line_plot, marker = '*')
plt.title("Line Plot")
plt.xlabel("Weeks")
plt.ylabel("Temperature")
plt.grid()
plt.show()
```

```python
scatter_plot = [1, 2, 3, 4, 5]
plt.scatter(scatter_plot, scatter_plot)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
pie_plot = [40, 30, 30]
activities = ["Work", "Sleep", "Leisure"]
plt.pie(pie_plot, labels = activities, autopct = '%0.1f%%')
plt.title("Pie Plot")
plt.grid()
plt.show()
```
