#### **Task 2: Combine and Split Arrays**

1. Create two 2D arrays of shape (3, 3) with random integers between 1 and 20.

```python
# Write your answer from this cell onwards
```

```python
array3_2d = np.random.randint(1, 21, size=(3, 3))
print(f"Array3: {array3_2d}")
array4_2d = np.random.randint(1, 21, size=(3, 3))
print(f"Array4: {array4_2d}")
```

2. Combine these arrays:
   - Vertically using `vstack`
   - Horizontally using `hstack`

```python
# Write your answer from this cell onwards
```

```python
vstack = np.vstack((array3_2d, array4_2d))
print(f"vstack: {vstack}")

hstack = np.hstack((array3_2d, array4_2d))
print(f"hstack: {hstack}")
```

3. Split the combined array back into the original two arrays using `vsplit` and `hsplit`.

```python
# Write your answer from this cell onwards
```

```python
vsplit = np.vsplit(vstack, 2)
print(f"vsplit: {vsplit}")

hsplit = np.hsplit(hstack, 2)
print(f"hsplit: {hsplit}")
```

#### **Task 5: Statistical Functions**

1. Create a 2D array of shape (5, 5) with random integers between 1 and 50.

```python
# Write your answer from this cell onwards
```

```python
array9 = np.random.randint(1, 50, size=(5, 5))
print(f"array9: {array9}")
array10 = np.random.randint(1, 50, size=(5, 5))
print(f"array10: {array10}")
```

2. Compute the following statistics:
   - Median of all elements.
   - Variance of the elements.
   - Cumulative sum along each row.
   - Cumulative product along each column.

```python
# Write your answer from this cell onwards
```

```python
meadian = np.median(array9)
print(f"meadian of array9: {meadian}")
meadian = np.median(array10)
print(f"meadian of array10: {meadian}")

variance = np.var(array9)
print(f"variance of array9: {variance}")
variance = np.var(array10)
print(f"variance of array10: {variance}")

cumulative_sum = np.cumsum(array9, axis=1)
print(f"cumulative_sum of array9: {cumulative_sum}")
cumulative_sum = np.cumsum(array10, axis=1)
print(f"cumulative_sum of array10: {cumulative_sum}")

cumulative_product = np.cumprod(array9, axis=0)
print(f"cumulative_product of array9: {cumulative_product}")
cumulative_product = np.cumprod(array10, axis=0)
print(f"cumulative_product of array10: {cumulative_product}")
```

#### **Task 6: Combine All Concepts**

1. Create a dataset using a 2D NumPy array with random integers between 1 and 100, where rows represent students and columns represent their scores in 5 subjects.

```python
# Write your answer from this cell onwards
```

```python
data_set = np.random.randint(1, 101, size=(10, 5))
print(f"data_set: {data_set}")
```

2. Perform the following:
   - Compute the total score for each student.
   - Find the student with the highest and lowest total score.
   - Sort the array based on total scores.
   - Filter students who scored more than 75 in at least three subjects.

```python
# Write your answer from this cell onwards
```

```python
total_score = np.sum(data_set, axis=1)
print(f"total_score: {total_score}")

highest_score = np.max(total_score)
print(f"highest_score: {highest_score}")

lowest_score = np.min(total_score)
print(f"lowest_score: {lowest_score}")

sorted_array = np.sort(total_score)
print(f"sorted_array: {sorted_array}")

filtered_array = data_set[np.where(np.sum(data_set[:, 3:] > 75, axis=1) >= 3)]
print(f"filtered_array: {filtered_array}")
```
