# Task 1: Combining DataFrames


1. Create two DataFrames, `df1` and `df2`, with overlapping columns but different data.
   - Use the `concat` method to vertically stack them.
   - Use the `merge` method to combine the DataFrames on a specific key column.
   - Use the `join` method to combine them based on their index.

```python
import pandas as pd
import numpy as np
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [2, 3, 4], 'C': [7, 8, 9]})
print(df1)
print(df2)
df_concat = pd.concat([df1, df2], axis=0)
df_merge = pd.merge(df1, df2, on='A')
df_join = df1.join(df2, lsuffix='_left', rsuffix='_right')
print(df_concat)
print(df_merge)
print(df_join)
```

2. Create two DataFrames with non-overlapping columns and combine them:
   - Horizontally stack them using `concat`.
   - Perform an outer join on a specific key column.

```python
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [2, 3, 4], 'C': [7, 8, 9]})
print(df1)
print(df2)
df_concat = pd.concat([df1, df2], axis=1)
df_merge = pd.merge(df1, df2, on='A', how='outer')
print(df_concat)
print(df_merge)
```

# Task 2: Handling Missing Data


1. Create a DataFrame with some missing values (`NaN`). Perform the following operations:
   - Fill missing values with a constant value.
   - Use `forward fill` and `backward fill` methods.
   - Replace missing values with the mean of the respective column.

```python
df1 = pd.DataFrame({'A': [1, 2, np.nan, 4], 'B': [5, np.nan, 7, 8]})
print(df1)
df_fill = df1.fillna(0)
print(df_fill)
df_ffill = df1.fillna(method='ffill')
print(df_ffill)
df_bfill = df1.fillna(method='bfill')
print(df_bfill)
df_mean = df1.fillna(df1.mean())
print(df_mean)
```

2. Identify and drop rows or columns with missing data:

```python
drop_row = df1.dropna(axis=0)
print(f"Original Row: {df1}")
print(drop_row)
print("\n--------------\n")
drop_col = df1.dropna(axis=1)
print(f"Original Column: {df1}")
print(drop_col)
```

3. Create a DataFrame with multiple groups of data (e.g., sales data for different regions):
   - Group the data by a specific column and calculate descriptive statistics (e.g., `mean`, `sum`, `std`) for each group.
   - Visualize summary statistics using a bar chart.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'West', 'West'],
        'Sales': [100, 150, 200, 250, 120, 180, 110, 160]}
df = pd.DataFrame(data)
grouped = df.groupby('Region')['Sales'].agg(['mean', 'sum', 'std'])
grouped.plot(kind='bar', figsize=(8, 6))
plt.title('Sales Summary by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.legend(title='Statistic')
plt.show()
```

# Additional Challenges


1. Combine DataFrames with mismatched data:
   - Resolve alignment issues by specifying `how` parameter in merging.

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df1 = pd.DataFrame({'A': [2, 3, 4], 'C': [7, 8, 9]})
print(df)
print(df1)
df_merge = pd.merge(df, df1, on='A', how='outer')
print(df_merge)
```

2. Generate a DataFrame with deliberate missing values and evaluate:
   - Impact of different imputation strategies (mean, median, mode).

```python
df = pd.DataFrame({'A': [1, 2, np.nan, 4], 'B': [5, np.nan, 7, 8]})
print(df)
print("-----------------\n")
df_mean = df.fillna(df.mean())
print(df_mean)
print("-----------------\n")
df_median = df.fillna(df.median())
print(df_median)
print("-----------------\n")
df_mode = df.fillna(df.mode().iloc[0])
print(df_mode)
```

3. Compare descriptive statistics before and after handling missing data in a dataset.

```python
sta = df.describe()
print(sta)
print("-----------------\n")
df_mean = df.fillna(df.mean())
sta_mean = df_mean.describe()
print(sta_mean)
print("-----------------\n")
df_median = df.fillna(df.median())
sta_median = df_median.describe()
print(sta_median)
print("-----------------\n")
df_mode = df.fillna(df.mode().iloc[0])
sta_mode = df_mode.describe()
print(sta_mode)
```
