# Task 1: Save a Chart Using Matplotlib

### Task: Create a bar chart showing the sales of different products and save it as 'sales_chart.png'.

### Steps:
1. Create a list of product names and their corresponding sales values.
2. Use Matplotlib to create a bar chart.
3. Save the chart using the `savefig` method.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = {'Product': ['A', 'B', 'C', 'D', 'E'],
        'Sales': [100, 150, 75, 200, 120]}
plt.bar(data['Product'], data['Sales'], color = 'yellow')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Sales of Different Products')
plt.savefig('sales_chart.png')
```

# Task 2: Data Visualization in Seaborn

### Task: Use Seaborn to visualize the relationship between 'sepal_length' and 'sepal_width' in the Iris dataset.

### Steps:
1. Load the Iris dataset using Seaborn.
2. Create a scatter plot using `sns.scatterplot`.
3. Customize the plot (e.g., add a title, labels).

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris
```

```python
sns.scatterplot(x = iris['sepal_length'],y = iris['sepal_width'])
plt.title('Relationship between Sepal Length and Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

# Task 3: Seaborn Project

### Task: Create a comprehensive dashboard using Seaborn visualizations.

### Project Steps:
1. Load the Titanic dataset from Seaborn.
2. Create the following plots:
   - A bar plot showing the survival count by gender.
   - A histogram of passenger ages.
   - A box plot of fare prices by class.
3. Customize each plot with appropriate titles and labels.


```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
plt.figure(figsize = (8,6))
sns.countplot(data = titanic, x = 'sex', hue = 'survived', palette = 'summer')
plt.title('Titanic outcomes')
plt.xlabel('Gender')
plt.ylabel('Counts')
plt.grid()
```

```python
plt.figure(figsize = (8,6))
sns.histplot(data = titanic, x = 'age', hue = 'who', palette = 'summer')
plt.title('Titanic outcomes')
plt.xlabel('Number of class')
plt.ylabel('Fares')
plt.grid()
```

```python
plt.figure(figsize = (8,6))
sns.boxplot(data = titanic, x = 'class', y = 'fare', color = 'pink')
plt.title('Titanic outcomes')
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.grid()
```
